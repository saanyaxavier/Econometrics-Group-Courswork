## 1.1 Data Preparation

```r
# Load libraries
install.packages("lmtest")
library(lmtest)
install.packages("writexl")
library(writexl)
library(sandwich)
library(ggplot2)
library(xts)
library(dplyr)
library(lubridate)
library(knitr)
library(data.table)
library(tidyverse)
library(tidyr)


# Load datasets
rates <- read.csv("Effective Federal Funds Rate (EFFR) (Monthly).csv")
ff3_raw <- read.csv("F-F_Research_Data_Factors.csv", skip = 3)
industry12_raw <- read.csv("12_Industry_Portfolios.csv", skip = 11)
industry49_raw <- read.csv("49_Industry_Portfolios.csv", skip = 11)
```

```r
head(rates)
glimpse(rates)
summary(rates)

head(ff3_raw)
glimpse(ff3_raw)

head(industry12_raw)
head(industry49_raw)
```

```r
# Clean data

# Clean rates
rates <- rates %>%
  rename(date = observation_date,
         interest_rate = EFFR)

glimpse(rates)

# Clean ff3
ff3 <- ff3_raw %>%
  rename(raw_date = `...1`) %>%
  filter(raw_date != "Annual") %>%
  mutate(
    date = as.Date(paste0(raw_date, "01"), format = "%Y%m%d"),
    market_excess_return = as.numeric(`Mkt-RF`) / 100
  ) %>%
  select(date, market_excess_return)


#Clean Industry 12
industry12 <- industry12_raw %>%
  rename(raw_date = `...1`) %>%
  filter(raw_date != "Annual") %>%
  mutate(
    date = as.Date(paste0(raw_date, "01"), format = "%Y%m%d")
  ) %>%
  select(-raw_date) %>%
  mutate(across(-date, as.numeric)) %>%
  mutate(across(-date, ~ . / 100))


# Clean Industry 49
toys <- industry49_raw %>%
  rename(raw_date = `...1`) %>%
  filter(raw_date != "Annual") %>%
  mutate(
    date = as.Date(paste0(raw_date, "01"), format = "%Y%m%d"),
    Toys = as.numeric(Toys) / 100
  ) %>%
  select(date, Toys)

head(ff3)
head(industry12)
head(toys)


# Merge datasets
data_merged <- ff3 %>%
  left_join(rates, by = "date") %>%
  left_join(toys, by = "date")


# Check merge quality
nrow(ff3) == nrow(data_merged)
sum(is.na(data_merged$interest_rate))
sum(is.na(data_merged$Toys))

glimpse(data_merged)

nrow(ff3) == nrow(data_merged)
sum(is.na(data_merged$interest_rate))
```
